### Recursos e Funções Sugeridas

Para completar este desafio, vocês podem usar diversas funções da biblioteca Pandas. Abaixo estão algumas que podem ser úteis, organizadas por categoria:

**Carregamento e Inspeção de Dados:**
*   `pd.read_csv()`: Para carregar dados de um arquivo CSV.
*   `df.head()`, `df.tail()`: Para visualizar as primeiras/últimas linhas do DataFrame.

**Operações Matemáticas e Lógicas:**

### Tarefa 2: Limpeza e Pré-processamento de Dados

1.  **Identifique** colunas com valores ausentes. Para cada coluna, mostre a contagem e a porcentagem de valores ausentes.
2.  **Preencha** os valores ausentes na coluna 'Age' com a mediana da idade.
3.  **Preencha** os valores ausentes na coluna 'Embarked' com o modo (valor mais frequente).
4.  **Descarte** a coluna 'Cabin' devido ao alto número de valores ausentes ou à sua complexidade para imputação simples neste desafio.
5.  **Converta** a coluna 'Sex' para valores numéricos (0 para 'male', 1 para 'female').
6.  **Crie** uma nova coluna 'FamilySize' que seja a soma de 'SibSp' (número de irmãos/cônjuges a bordo) e 'Parch' (número de pais/filhos a bordo), mais 1 (para o próprio passageiro).
7.  **Crie** uma nova coluna 'IsAlone' baseada em 'FamilySize' (1 se FamilySize for 1, 0 caso contrário).
8.  **Exiba** novamente `titanic_df.info()` e `titanic_df.head()` para verificar as alterações.

```python
#1 - Identifique colunas com valores ausentes. Para cada coluna, mostre a contagem e a porcentagem de valores ausentes.
titanic_df.isnull().sum()
```

```python
#2 Preencha os valores ausentes na coluna 'Age' com a mediana da idade.
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].median())
titanic_df.isnull().sum()
```

```python
#3 Preencha os valores ausentes na coluna 'Embarked' com o modo (valor mais frequente).
titanic_df['Embarked'] = titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0])
titanic_df.isnull().sum()
```

```python
#4 - Descarte a coluna 'Cabin' devido ao alto número de valores ausentes ou à sua complexidade para imputação simples neste desafio.
titanic_df = titanic_df.drop(columns=['Cabin'])
titanic_df.isnull().sum()
```

```python
#5 Converta a coluna 'Sex' para valores numéricos (0 para 'male', 1 para 'female').
print(titanic_df['Sex'].value_counts())
titanic_df['Sex'] = titanic_df['Sex'].replace({'male': 0, 'female': 1})
titanic_df.head()
```

```python
#6 Crie uma nova coluna 'FamilySize' que seja a soma de 'SibSp'
 #(número de irmãos/cônjuges a bordo) e 'Parch' (número de pais/filhos a bordo), mais 1 (para o próprio passageiro).

titanic_df['FamilySize'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1
titanic_df.head()
```

```python
#7 Crie uma nova coluna 'IsAlone' baseada em 'FamilySize' (1 se FamilySize for 1, 0 caso contrário).
titanic_df['IsAlone'] = (titanic_df['FamilySize'] == 1).astype(int)
titanic_df.head()
```

```python
#8 Exiba novamente titanic_df.info() e titanic_df.head() para verificar as alterações.
titanic_df.info()
titanic_df.head()
```

### Tarefa 3: Análise Exploratória de Dados (EDA) com Pandas

1.  **Qual** a proporção de sobreviventes no total?
2.  **Calcule** a taxa de sobrevivência por gênero.
3.  **Calcule** a taxa de sobrevivência por classe de passageiro ('Pclass').
4.  **Determine** a idade média dos sobreviventes e dos não-sobreviventes.
5.  **Encontre** a porta de embarque (Embarked) mais comum e a taxa de sobrevivência para cada porta.
6.  **Crie** uma tabela dinâmica (pivot table) mostrando a taxa de sobrevivência média por 'Pclass' e 'Sex'.

```python
#1 Qual a proporção de sobreviventes no total?
titanic_df['Survived'].value_counts(normalize=True)*100
```

```python
#2 Calcule a taxa de sobrevivência por gênero.
genero_sobre = titanic_df.groupby('Sex')['Survived'].mean()*100
genero_sobre
```

```python
#3 Calcule a taxa de sobrevivência por classe de passageiro ('Pclass').
classe_sobre = titanic_df.groupby('Pclass')['Survived'].mean()
classe_sobre
```

```python
#4 Determine a idade média dos sobreviventes e dos não-sobreviventes.
idade_sobre = titanic_df.groupby('Survived')['Age'].mean()
idade_sobre
```

```python
#5 Encontre a porta de embarque (Embarked) mais comum e a taxa de sobrevivência para cada porta.
embarked_moda = titanic_df['Embarked'].mode()[0]
print(f'Porta de embarque mais comum: {embarked_moda}')

embarked_sobre = titanic_df.groupby('Embarked')['Survived'].mean()
embarked_sobre
```

```python
#6 Crie uma tabela dinâmica (pivot table) mostrando a taxa de sobrevivência média por 'Pclass' e 'Sex'.
tabela_dinamica = titanic_df.pivot_table(
    values='Survived',
    index='Pclass',
    columns='Sex',
    aggfunc='mean'
)
print("Taxa de Sobrevivência Média por Classe e Gênero:")
tabela_dinamica
```

### Tarefa 4: Manipulações Avançadas com Pandas

1.  **Crie** uma coluna 'AgeGroup' categorizando a idade em faixas: 'Criança' (0-12), 'Adolescente' (13-18), 'Adulto' (19-59), 'Idoso' (60+). Use `pd.cut`.
2.  **Encontre** a taxa de sobrevivência para cada 'AgeGroup'.
3.  **Identifique** os 5 nomes de passageiros mais comuns no dataset (após limpeza, se aplicável, ou apenas a coluna 'Name'). Nota: Isso pode ser complicado devido à variação de nomes. Uma abordagem pode ser extrair apenas o sobrenome ou o título.
4.  **Conte** a frequência de cada título (e.g., Mr., Miss., Mrs., Master.) na coluna 'Name'. Você pode extrair o título usando expressões regulares ou manipulação de strings. Ex: `df['Name'].str.extract(' ([A-Za-z]+)\.')`.
5.  **Calcule** a tarifa ('Fare') média por 'Pclass' e 'Embarked'.

```python
#1 Crie uma coluna 'AgeGroup' categorizando a idade em faixas: 'Criança' (0-12), 'Adolescente' (13-18), 'Adulto' (19-59), 'Idoso' (60+). Use pd.cut.
import pandas as pd
import numpy as np

#(bins)
bins = [0, 12, 18, 59, np.inf]

#categorias
labels = ['Criança', 'Adolescente', 'Adulto', 'Idoso']

titanic_df['AgeGroup'] = pd.cut(titanic_df['Age'], bins=bins, labels=labels)

titanic_df[['Age', 'AgeGroup']].head(20)
```

```python
#2 Encontre a taxa de sobrevivência para cada 'AgeGroup'.
agegroup_sobre = titanic_df.groupby('AgeGroup')['Survived'].mean()
agegroup_sobre
```

```python
#3 Identifique os 5 nomes de passageiros mais comuns no dataset (após limpeza, se aplicável, ou apenas a coluna 'Name').
#Nota: Isso pode ser complicado devido à variação de nomes. Uma abordagem pode ser extrair apenas o sobrenome ou o título.

# Extraindo o sobrenome (tudo que vem antes da vírgula)
titanic_df['Surname'] = titanic_df['Name'].str.split(',').str[0]

#  5 primeiros
top_5_sobrenomes = titanic_df['Surname'].value_counts().head(5)

print("Os 5 sobrenomes mais comuns:")
top_5_sobrenomes
```

```python
#4 Conte a frequência de cada título (e.g., Mr., Miss., Mrs., Master.) na coluna 'Name'.
#Você pode extrair o título usando expressões regulares ou manipulação de strings. Ex: df['Name'].str.extract(' ([A-Za-z]+)\.').

# Ele procura um espaço, seguido de letras, terminando com um ponto
titanic_df['Title'] = titanic_df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

#frequentes
top_5_titulos = titanic_df['Title'].value_counts().head(5)

print("\nOs 5 títulos mais comuns:")
top_5_titulos
```

```python
#5 Calcule a tarifa ('Fare') média por 'Pclass' e 'Embarked'.
tarifa_media = titanic_df.groupby(['Pclass', 'Embarked'])['Fare'].mean()
tarifa_media
```
