```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
dados = pd.DataFrame(
    {
        'X': [1,2,3,4,5],
        'Y': [1.3, 1.8,3.5,4,4.6]
    }
)
```

```python
dados.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

ax.scatter(dados.X, dados.Y)
ax.plot(dados.X, dados.Y, '--r')
plt.scatter(dados.X, dados.Y)

plt.show()

```

```python
dados['y_reta'] = dados.X
```

```python
dados
```

```python
fig, ax = plt.subplots()

ax.scatter(dados.X, dados.y_reta)
ax.plot(dados.X, dados.y_reta, '--r')
plt.scatter(dados.X, dados.y_reta)

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(dados.X.values.reshape(-1,1), dados.Y)
```

```python
a = reg.coef_[0]
a
```

```python
b = reg.intercept_
b
```

```python
#Analisar a reta que o modelo descobriu que melhor representa nosso dados

fig, ax = plt.subplots()

ax.scatter(dados.X, dados.Y)
ax.plot(dados.X, dados.y_reta, '--r')

x = dados.X.values
y = a*x + b
ax.plot(x, y)

plt.show()
```

```python
#Fazendo previsão e adicionando no nosso DataFrame
dados['y_pred'] = reg.predict(dados.X.values.reshape(-1,1))
```

```python
dados
```

```python
#y real e y predito para nossa reta e para reta do modelo
dados['erra_reta'] = (dados.Y - dados.y_reta)**2
dados['erro_pred'] = (dados.Y - dados.y_pred)**2
```

```python
dados
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_absolute_error(dados.Y, dados.y_reta))
print(mean_absolute_error(dados.Y, dados.y_pred))
print(mean_squared_error(dados.Y, dados.y_reta))
print(mean_squared_error(dados.Y, dados.y_pred))
```
