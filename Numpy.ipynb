 ## 1. Listas

### Teoria Básica

Listas em Python são coleções ordenadas e mutáveis (ou seja, podem ser alteradas) de itens. Elas podem armazenar elementos de diferentes tipos de dados (inteiros, strings, floats, outras listas, etc.). São definidas usando colchetes `[]` e os itens são separados por vírgulas.

````python
# Exemplo de uso de Listas

# Criando uma lista
minha_lista = [1, 'olá', 3.14, True]
print(f"Lista original: {minha_lista}")

# Acessando elementos (índices começam em 0)
print(f"Primeiro elemento: {minha_lista[0]}")
print(f"Último elemento: {minha_lista[-1]}")

# Modificando elementos
minha_lista[1] = 'mundo'
print(f"Lista após modificação: {minha_lista}")

# Adicionando elementos
minha_lista.append(5)
print(f"Lista após append(5): {minha_lista}")

minha_lista.insert(1, 'novo item') # Insere 'novo item' no índice 1
print(f"Lista após insert(1, 'novo item'): {minha_lista}")

# Removendo elementos
minha_lista.remove(True) # Remove a primeira ocorrência do valor True
print(f"Lista após remove(True): {minha_lista}")

item_removido = minha_lista.pop(0) # Remove e retorna o item no índice 0
print(f"Item removido com pop(0): {item_removido}")
print(f"Lista após pop(0): {minha_lista}")

# Concatenando listas
lista_extra = [10, 20]
lista_concatenada = minha_lista + lista_extra
print(f"Lista concatenada: {lista_concatenada}")

# Estendendo uma lista com outra
minha_lista.extend([6, 7])
print(f"Lista após extend([6, 7]): {minha_lista}")

# Verificando o tamanho da lista
print(f"Tamanho da lista: {len(minha_lista)}")

# Ordenando uma lista (se os elementos forem comparáveis)
numeros = [3, 1, 4, 1, 5, 9, 2]
numeros.sort()
print(f"Lista ordenada: {numeros}")

# Invertendo a ordem de uma lista
numeros.reverse()
print(f"Lista invertida: {numeros}")

# Contando ocorrências de um elemento
print(f"Número de '1' na lista: {numeros.count(1)}")
````

 ## 2. Dicionários

### Teoria Básica

Dicionários em Python são coleções não ordenadas e mutáveis de pares `chave: valor`. Cada chave deve ser única e imutável (como strings, números ou tuplas), e os valores podem ser de qualquer tipo de dado. Dicionários são definidos usando chaves `{}` e os pares são separados por vírgulas.

````python
# Exemplo de uso de Dicionários

# Criando um dicionário
dicionario_aluno = {
    'nome': 'Ana',
    'idade': 20,
    'curso': 'Engenharia',
    'notas': [8.5, 9.0, 7.8]
}
print(f"Dicionário original: {dicionario_aluno}")

# Acessando valores
print(f"Nome do aluno: {dicionario_aluno['nome']}")
print(f"Idade do aluno (usando get): {dicionario_aluno.get('idade')}")
print(f"Curso do aluno (chave inexistente com get, valor padrão): {dicionario_aluno.get('turma', 'Não informado')}")

# Modificando valores
dicionario_aluno['idade'] = 21
print(f"Dicionário após modificar idade: {dicionario_aluno}")

# Adicionando novos pares chave:valor
dicionario_aluno['cidade'] = 'São Paulo'
print(f"Dicionário após adicionar cidade: {dicionario_aluno}")

# Removendo pares chave:valor
curso_removido = dicionario_aluno.pop('curso') # Remove e retorna o valor associado à chave
print(f"Curso removido: {curso_removido}")
print(f"Dicionário após remover curso: {dicionario_aluno}")

del dicionario_aluno['notas'] # Remove o par 'notas': [...] usando del
print(f"Dicionário após remover notas com del: {dicionario_aluno}")

# Obtendo todas as chaves
print(f"Chaves do dicionário: {dicionario_aluno.keys()}")

# Obtendo todos os valores
print(f"Valores do dicionário: {dicionario_aluno.values()}")

# Obtendo todos os itens (pares chave:valor)
print(f"Itens do dicionário: {dicionario_aluno.items()}")

# Verificando o tamanho do dicionário
print(f"Número de pares no dicionário: {len(dicionario_aluno)}")

# Atualizando um dicionário com outro
dados_complementares = {'estado': 'SP', 'matricula': '2023001'}
dicionario_aluno.update(dados_complementares)
print(f"Dicionário após update: {dicionario_aluno}")
````

 ## 3. Tuplas

### Teoria Básica

Tuplas em Python são coleções ordenadas e **imutáveis** de itens. Isso significa que, uma vez criada, uma tupla não pode ter seus elementos adicionados, removidos ou modificados. Elas podem armazenar elementos de diferentes tipos de dados. São definidas usando parênteses `()` e os itens são separados por vírgulas. A imutabilidade torna as tuplas ideais para dados que não devem mudar, como coordenadas ou registros fixos.

````python
# Exemplo de uso de Tuplas

# Criando uma tupla
minha_tupla = (10, 'dez', 20.5, False)
print(f"Tupla original: {minha_tupla}")

# Tupla com um único elemento (note a vírgula)
tupla_unitaria = (5,)
print(f"Tupla unitária: {tupla_unitaria}, tipo: {type(tupla_unitaria)}")

# Acessando elementos (índices começam em 0)
print(f"Primeiro elemento: {minha_tupla[0]}")
print(f"Terceiro elemento: {minha_tupla[2]}")

# Tentativa de modificar um elemento (resultará em erro)
try:
    minha_tupla[1] = 'vinte'
except TypeError as e:
    print(f"Erro ao tentar modificar tupla: {e}")

# Concatenando tuplas (cria uma nova tupla)
tupla_extra = ('a', 'b')
tupla_concatenada = minha_tupla + tupla_extra
print(f"Tupla concatenada: {tupla_concatenada}")

# Verificando o tamanho da tupla
print(f"Tamanho da tupla: {len(minha_tupla)}")

# Contando ocorrências de um elemento
tupla_numeros = (1, 2, 2, 3, 4, 2)
print(f"Número de '2' na tupla: {tupla_numeros.count(2)}")

# Encontrando o índice da primeira ocorrência de um elemento
print(f"Índice da primeira ocorrência de '3': {tupla_numeros.index(3)}")

# Desempacotamento de tuplas
coordenadas = (10.0, 20.0)
x, y = coordenadas
print(f"Coordenada X: {x}, Coordenada Y: {y}")

# Tuplas podem ser usadas como chaves em dicionários (por serem imutáveis)
dicionario_com_tupla_chave = {
    (1, 2): 'Ponto A',
    (3, 4): 'Ponto B'
}
print(f"Valor para a chave (1, 2): {dicionario_com_tupla_chave[(1, 2)]}")
````

 ## 4. NumPy

### Teoria Básica

NumPy (Numerical Python) é uma biblioteca fundamental para computação científica em Python. Ela fornece um objeto de array multidimensional de alto desempenho, e ferramentas para trabalhar com esses arrays. O NumPy é a base para muitas outras bibliotecas científicas em Python, como SciPy, Matplotlib e scikit-learn.

#### Para que usar?

*   **Arrays Multidimensionais (ndarray)**: Permite armazenar e manipular grandes coleções de dados numéricos de forma eficiente.
*   **Funções Matemáticas para Arrays**: Oferece uma vasta coleção de funções matemáticas para operar em arrays inteiros, sem a necessidade de loops explícitos em Python, o que resulta em um desempenho muito superior.
*   **Computação Numérica Eficiente**: Implementado em C e Fortran, o NumPy é significativamente mais rápido que as listas Python para operações numéricas em grandes volumes de dados.
*   **Broadcasting**: Um mecanismo que permite que o NumPy opere em arrays com diferentes formas durante operações aritméticas.
*   **Vetorização**: A capacidade de aplicar operações a um array inteiro de uma vez, em vez de elemento por elemento, melhorando o desempenho e a legibilidade do código.

  ### Arrays Multidimensionais

O objeto central do NumPy é o `ndarray`, que significa "N-dimensional array". É um contêiner homogêneo de elementos (todos os elementos devem ser do mesmo tipo de dado).

*   **Criação de Arrays**: Pode ser feito a partir de listas Python ou usando funções como [`np.zeros`](https://numpy.org/doc/stable/reference/generated/numpy.zeros.html), [`np.ones`](https://numpy.org/doc/stable/reference/generated/numpy.ones.html), [`np.arange`](https://numpy.org/doc/stable/reference/generated/numpy.arange.html), [`np.linspace`](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html), [`np.random.rand`](https://numpy.org/doc/stable/reference/random/generated/numpy.random.rand.html), etc.
*   **Acesso e Indexação**: Similar às listas Python, mas com capacidades de indexação avançadas.
    *   **Slicing (Fatiamento)**: Permite extrair subarrays ou subconjuntos de elementos usando a notação `[start:stop:step]`, similar às listas Python.
    *   **Indexação Booleana**: Utiliza um array booleano (verdadeiro/falso) para selecionar elementos que satisfazem uma determinada condição. A forma do array booleano deve ser a mesma do array original ou broadcasting compatível.
    *   **Fancy Indexing (Indexação Avançada)**: Permite selecionar elementos ou linhas/colunas arbitrárias usando arrays de índices inteiros. Isso cria uma *cópia* dos dados, ao contrário do slicing que retorna uma *view*.
*   **Atributos Importantes**: [`shape`](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.shape.html) (dimensões do array), [`ndim`](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.ndim.html) (número de dimensões), [`size`](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.size.html) (número total de elementos), [`dtype`](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.dtype.html) (tipo de dado dos elementos).


````python
# Criando um array 1D a partir de uma lista
arr1d = np.array([1, 2, 3, 4, 5])
print(f"Array 1D: {arr1d}")
print(f"Shape do array 1D: {arr1d.shape}") # (5,)
print(f"Dimensões do array 1D: {arr1d.ndim}") # 1

# Criando um array 2D (matriz) a partir de uma lista de listas
arr2d = np.array([[1, 2, 3], [4, 5, 6]])
print(f"\nArray 2D:\n{arr2d}")
print(f"Shape do array 2D: {arr2d.shape}") # (2, 3)
print(f"Dimensões do array 2D: {arr2d.ndim}") # 2

# Criando um array de zeros
zeros_array = np.zeros((3, 4))
print(f"\nArray de zeros (3x4):\n{zeros_array}")

# Criando um array com valores sequenciais
ranged_array = np.arange(0, 10, 2) # Start, Stop (exclusive), Step
print(f"\nArray com arange: {ranged_array}")

# Acessando elementos
print(f"\nPrimeiro elemento do arr1d: {arr1d[0]}")
print(f"Elemento na linha 0, coluna 1 do arr2d: {arr2d[0, 1]}")

# Slicing (fatiamento)
print(f"Primeiras 3 elementos do arr1d: {arr1d[:3]}")
print(f"Elementos do arr1d do índice 1 ao 3: {arr1d[1:4]}")
print(f"Primeira linha do arr2d: {arr2d[0, :]}")
print(f"Segunda coluna do arr2d: {arr2d[:, 1]}")
print(f"Submatriz de arr2d: {arr2d[0:1, 1:3]}")

# Indexação Booleana
arr_bool = np.array([10, 20, 30, 40, 50, 60])
condicao = arr_bool > 30
print(f"\nArray para indexação booleana: {arr_bool}")
print(f"Condição (arr_bool > 30): {condicao}")
print(f"Elementos maiores que 30: {arr_bool[condicao]}")

# Fancy Indexing
arr_fancy = np.array(['a', 'b', 'c', 'd', 'e'])
indices = [0, 2, 4]
print(f"\nArray para fancy indexing: {arr_fancy}")
print(f"Elementos nos índices {indices}: {arr_fancy[indices]}")

matrix_fancy = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])
linhas_selecionadas = np.array([0, 2])
colunas_selecionadas = np.array([1, 0])
print(f"\nMatriz para fancy indexing:\n{matrix_fancy}")
print(f"Elementos específicos (linhas, colunas): {matrix_fancy[linhas_selecionadas, colunas_selecionadas]}") # Retorna (matrix_fancy[0,1], matrix_fancy[2,0])
````

  ### Computação Numérica

O NumPy oferece um desempenho superior para operações matemáticas em arrays em comparação com as listas Python. Ele permite realizar operações elemento a elemento de forma eficiente.

*   **Operações Aritméticas Básicas**: `+`, `-`, `*`, `/`, `**` (potência) aplicadas elemento a elemento.
*   **Funções Universais (ufuncs)**: Funções como [`np.sqrt()`](https://numpy.org/doc/stable/reference/generated/numpy.sqrt.html), [`np.sin()`](https://numpy.org/doc/stable/reference/generated/numpy.sin.html), [`np.log()`](https://numpy.org/doc/stable/reference/generated/numpy.log.html), [`np.exp()`](https://numpy.org/doc/stable/reference/generated/numpy.exp.html) que operam elemento a elemento.
*   **Funções de Agregação**: [`np.sum()`](https://numpy.org/doc/stable/reference/generated/numpy.sum.html), [`np.mean()`](https://numpy.org/doc/stable/reference/generated/numpy.mean.html), [`np.min()`](https://numpy.org/doc/stable/reference/generated/numpy.min.html), [`np.max()`](https://numpy.org/doc/stable/reference/generated/numpy.max.html), [`np.std()`](https://numpy.org/doc/stable/reference/generated/numpy.std.html) para calcular estatísticas sobre arrays inteiros ou ao longo de um eixo específico.


````python
# Criando dois arrays para demonstração
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

# Operações aritméticas elemento a elemento
print(f"Soma: {a + b}")
print(f"Subtração: {a - b}")
print(f"Multiplicação: {a * b}")
print(f"Divisão: {a / b}")
print(f"Potência: {a ** 2}") # Eleva cada elemento de 'a' ao quadrado

# Funções universais (ufuncs)
print(f"Raiz quadrada de 'a': {np.sqrt(a)}")
print(f"Seno de 'b': {np.sin(b)}")

# Funções de agregação
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(f"\nMatrix:\n{matrix}")
print(f"Soma de todos os elementos: {np.sum(matrix)}")
print(f"Média de todos os elementos: {np.mean(matrix):.2f}")
print(f"Soma ao longo do eixo 0 (colunas): {np.sum(matrix, axis=0)}")
print(f"Soma ao longo do eixo 1 (linhas): {np.sum(matrix, axis=1)}")
````

 ### Broadcasting

Broadcasting é o termo que descreve como o NumPy trata arrays com diferentes shapes (formas) durante operações aritméticas. Sujeito a certas restrições, o array menor é "transmitido" (broadcast) ao array maior para que eles tenham shapes compatíveis.

As regras de broadcasting são:

1.  Se os arrays não tiverem o mesmo número de dimensões, a forma do array com menos dimensões é prefixada com uns.
2.  Se a forma dos arrays não coincidir em nenhuma dimensão, o array com a forma igual a 1 nessa dimensão é esticado para corresponder à outra forma.
3.  Se em alguma dimensão os tamanhos forem diferentes e nenhum for 1, ocorre um erro (ValueError).

````python
# Array 1D e escalar
arr = np.array([1, 2, 3])
scalar = 2
print(f"Array: {arr}")
print(f"Escalar: {scalar}")
print(f"Array + Escalar (broadcasting): {arr + scalar}") # 2 é broadcast para [2, 2, 2]

# Array 2D e array 1D
A = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([10, 20, 30])
print(f"\nMatriz A:\n{A}")
print(f"Vetor b: {b}")
print(f"A + b (broadcasting):\n{A + b}") # b é broadcast para [[10, 20, 30], [10, 20, 30]]

# Array 2D e array 1D (coluna)
c = np.array([[10], [20]])
print(f"\nMatriz A:\n{A}")
print(f"Vetor c (coluna):\n{c}")
print(f"A + c (broadcasting):\n{A + c}") # c é broadcast para [[10, 10, 10], [20, 20, 20]]
````

 ### Vetorização

Vetorização é o processo de escrever operações em arrays de forma a eliminar loops explícitos. Em vez de iterar sobre os elementos de um array um por um, as operações são aplicadas a todo o array de uma vez. Isso é crucial para o desempenho no NumPy, pois as operações vetorizadas são implementadas em C e são muito mais rápidas do que os loops em Python.

*   **Benefícios**: Melhor desempenho, código mais limpo e conciso, e menor chance de erros.

````python
# Exemplo de operação não vetorizada (usando loop Python)
lista = [1, 2, 3, 4, 5]
resultado_loop = []
import time

start_time = time.time()
for x in lista:
    resultado_loop.append(x * 2)
end_time = time.time()
print(f"Resultado com loop: {resultado_loop}")
print(f"Tempo com loop: {end_time - start_time:.6f} segundos")

# Exemplo de operação vetorizada com NumPy
numpy_array = np.array([1, 2, 3, 4, 5])

start_time = time.time()
resultado_vetorizado = numpy_array * 2
end_time = time.time()
print(f"\nResultado vetorizado: {resultado_vetorizado}")
print(f"Tempo vetorizado: {end_time - start_time:.6f} segundos")

# Demonstração em um cenário maior
large_array = np.random.rand(1_000_000)
large_list = large_array.tolist()

# Tempo para loop Python
start_time = time.time()
result_list = []
for x in large_list:
    result_list.append(np.sqrt(x))
end_time = time.time()
print(f"\nTempo para raiz quadrada com loop em lista de 1M elementos: {end_time - start_time:.6f} segundos")

# Tempo para operação vetorizada com NumPy
start_time = time.time()
result_numpy = np.sqrt(large_array)
end_time = time.time()
print(f"Tempo para raiz quadrada vetorizada em array de 1M elementos: {end_time - start_time:.6f} segundos")

# A diferença de desempenho é clara em grandes datasets.
````

  ## 6. Sorting, Searching e Counting com NumPy

### Teoria Básica

NumPy oferece funções eficientes para ordenar, buscar e contar elementos em arrays, que são essenciais para análise e manipulação de dados.

*   [`np.sort()`](https://numpy.org/doc/stable/reference/generated/numpy.sort.html): Retorna uma cópia ordenada do array. Pode ordenar por eixos específicos em arrays multidimensionais.
*   [`np.argsort()`](https://numpy.org/doc/stable/reference/generated/numpy.argsort.html): Retorna os *índices* que ordenariam um array. Útil para ordenar múltiplos arrays com base na ordem de um.
*   [`np.where(condition, x, y)`](https://numpy.org/doc/stable/reference/generated/numpy.where.html): Retorna elementos selecionados de `x` ou `y` dependendo da `condition`. Se `x` e `y` não forem fornecidos, retorna os índices onde a condição é verdadeira.
*   [`np.unique()`](https://numpy.org/doc/stable/reference/generated/numpy.unique.html): Retorna os elementos únicos de um array ordenado. Pode retornar também os índices dos elementos únicos, os índices inversos e as contagens.
*   [`np.count_nonzero()`](https://numpy.org/doc/stable/reference/generated/numpy.count_nonzero.html): Conta o número de elementos não-zero em um array.


````python
import numpy as np

# Exemplo de Sorting, Searching e Counting

arr_dados = np.array([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5])

print(f"Array original: {arr_dados}")

# Sorting
arr_ordenado = np.sort(arr_dados)
print(f"Array ordenado: {arr_ordenado}")

matriz_sort = np.array([[5, 2, 8], [1, 9, 4]])
print(f"\nMatriz para sort:\n{matriz_sort}")
print(f"Matriz ordenada por linha (axis=1):\n{np.sort(matriz_sort, axis=1)}")
print(f"Matriz ordenada por coluna (axis=0):\n{np.sort(matriz_sort, axis=0)}")

# Argsort
indices_ordenacao = np.argsort(arr_dados)
print(f"\nÍndices que ordenariam o array: {indices_ordenacao}")
print(f"Array ordenado usando argsort: {arr_dados[indices_ordenacao]}")

# Where
indices_maior_que_5 = np.where(arr_dados > 5)
print(f"\nÍndices onde os elementos são maiores que 5: {indices_maior_que_5}")
print(f"Elementos maiores que 5: {arr_dados[indices_maior_que_5]}")

substituicao = np.where(arr_dados > 5, 99, arr_dados)
print(f"Array com elementos maiores que 5 substituídos por 99: {substituicao}")

# Unique
elementos_unicos = np.unique(arr_dados)
print(f"\nElementos únicos: {elementos_unicos}")

elementos_unicos_contagem = np.unique(arr_dados, return_counts=True)
print(f"Elementos únicos e suas contagens: {elementos_unicos_contagem}")

# Count Nonzero
arr_zero = np.array([0, 1, 0, 5, 0, -2])
print(f"\nArray para contar não-zeros: {arr_zero}")
print(f"Número de elementos não-zero: {np.count_nonzero(arr_zero)}")
````

  ## 7. Cópia de Arrays (Views vs. Copies) no NumPy

### Teoria Básica

No NumPy, é crucial entender a diferença entre uma 'view' e uma 'copy' ao manipular arrays, pois isso afeta como as modificações são propagadas.

*   **Copy (Cópia)**: Cria um novo array e um novo espaço na memória. Modificações na cópia não afetam o array original, e vice-versa. O método [`copy()`](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.copy.html) explicitamente cria uma cópia.
*   **View (Visão)**: Cria um novo objeto array que *aponta para os mesmos dados* do array original. Modificações na view **afetam** o array original, e vice-versa. Slicing e algumas operações de reshape geralmente retornam views. O método [`view()`](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.view.html) explicitamente cria uma view.

A regra geral é: se você deseja modificar uma parte de um array sem alterar o original, use `.copy()`. Se você quer economizar memória e as modificações devem ser refletidas no original (ou vice-versa), uma view é apropriada.


````python
import numpy as np

# Exemplo de Views vs. Copies

# Array Original
arr_original = np.array([1, 2, 3, 4, 5])
print(f"Array Original: {arr_original}")

# --- Criando uma View (Slicing) ---
view_array = arr_original[1:4] # Slicing geralmente cria uma view
print(f"View do array (slice): {view_array}")

# Modificando a view
view_array[0] = 99
print(f"View após modificação: {view_array}")
print(f"Array Original após modificação da View: {arr_original}") # O original é alterado!

# Verificando se os dados são os mesmos
print(f"São os mesmos dados? (base da view é o original): {view_array.base is arr_original}")

# --- Criando uma Cópia (Explicitamente com .copy()) ---
arr_original_2 = np.array([10, 20, 30, 40, 50])
copy_array = arr_original_2.copy() # Cria uma cópia explícita
print(f"\nArray Original 2: {arr_original_2}")
print(f"Cópia do array: {copy_array}")

# Modificando a cópia
copy_array[0] = 111
print(f"Cópia após modificação: {copy_array}")
print(f"Array Original 2 após modificação da Cópia: {arr_original_2}") # O original NÃO é alterado!

# Verificando se os dados são os mesmos
print(f"São os mesmos dados? (base da cópia é o original): {copy_array.base is arr_original_2}") # Deve ser False ou None
````

  ## 8. Manipulação de Shape (Reshape, Transpose) no NumPy

### Teoria Básica

A manipulação da forma (shape) dos arrays é uma operação fundamental no NumPy, especialmente quando se trabalha com modelos de Machine Learning que exigem entradas com dimensões específicas. As funções mais comuns são:

*   [`reshape(new_shape)`](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.reshape.html): Retorna um array com o mesmo número de elementos, mas com uma nova forma. O número total de elementos deve ser consistente.
*   [`transpose()`](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.transpose.html) ou [`T`](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.T.html): Inverte ou permuta os eixos de um array. Para matrizes 2D, isso significa trocar linhas por colunas.
*   [`flatten()`](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.flatten.html): Retorna uma cópia do array recolhido em 1D.
*   [`ravel()`](https://numpy.org/doc/stable/reference/generated/numpy.ravel.html): Retorna uma view (se possível) ou uma cópia do array recolhido em 1D. Mais eficiente que `flatten` se uma view for suficiente.
*   [`np.newaxis`](https://numpy.org/doc/stable/reference/constants.html#numpy.newaxis): Usado para aumentar as dimensões de um array, adicionando um novo eixo (por exemplo, para transformar um vetor 1D em um vetor coluna 2D).


````python
import numpy as np

# Exemplo de Manipulação de Shape

arr_1d = np.arange(12)
print(f"Array 1D: {arr_1d}, Shape: {arr_1d.shape}")

# Reshape para 2D (3 linhas, 4 colunas)
arr_2d_3x4 = arr_1d.reshape(3, 4)
print(f"\nArray 2D (3x4):\n{arr_2d_3x4}, Shape: {arr_2d_3x4.shape}")

# Reshape para 2D (4 linhas, 3 colunas)
arr_2d_4x3 = arr_1d.reshape(4, 3)
print(f"Array 2D (4x3):\n{arr_2d_4x3}, Shape: {arr_2d_4x3.shape}")

# Reshape com -1 (NumPy calcula a dimensão automaticamente)
arr_2d_auto = arr_1d.reshape(2, -1) # 2 linhas, colunas calculadas (6)
print(f"Array 2D (2 linhas, colunas auto):\n{arr_2d_auto}, Shape: {arr_2d_auto.shape}")

# Transpose (para matriz 2D)
matriz_original = np.array([[1, 2, 3], [4, 5, 6]])
print(f"\nMatriz Original:\n{matriz_original}, Shape: {matriz_original.shape}")

matriz_transposta = matriz_original.T # ou matriz_original.transpose()
print(f"Matriz Transposta:\n{matriz_transposta}, Shape: {matriz_transposta.shape}")

# Flatten vs Ravel
flat_copy = matriz_original.flatten()
raveled_view_or_copy = matriz_original.ravel()
print(f"\nArray Original (para flatten/ravel):\n{matriz_original}")
print(f"Flatten (cópia 1D): {flat_copy}, Shape: {flat_copy.shape}")
print(f"Ravel (view/cópia 1D): {raveled_view_or_copy}, Shape: {raveled_view_or_copy.shape}")

# Modificando ravel para ver se é view
raveled_view_or_copy[0] = 999
print(f"Matriz Original após modificar ravel: {matriz_original}") # Original foi alterado, então ravel retornou uma view

# np.newaxis para adicionar dimensão
vetor = np.array([1, 2, 3])
vetor_linha = vetor[np.newaxis, :]
vetor_coluna = vetor[:, np.newaxis]
print(f"\nVetor original: {vetor}, Shape: {vetor.shape}")
print(f"Vetor como linha (1x3): {vetor_linha}, Shape: {vetor_linha.shape}")
print(f"Vetor como coluna (3x1): {vetor_coluna}, Shape: {vetor_coluna.shape}")
````

  ## 9. Criação de Arrays Específicos com NumPy

### Teoria Básica

NumPy oferece várias funções convenientes para criar arrays com padrões específicos, além de converter listas Python. Essas funções são úteis para inicializar arrays, criar dados de teste ou para operações matemáticas específicas.

*   [`np.zeros(shape)`](https://numpy.org/doc/stable/reference/generated/numpy.zeros.html): Cria um array preenchido com zeros.
*   [`np.ones(shape)`](https://numpy.org/doc/stable/reference/generated/numpy.ones.html): Cria um array preenchido com uns.
*   [`np.full(shape, fill_value)`](https://numpy.org/doc/stable/reference/generated/numpy.full.html): Cria um array preenchido com um valor específico.
*   [`np.empty(shape)`](https://numpy.org/doc/stable/reference/generated/numpy.empty.html): Cria um array com dados aleatórios (não inicializados), mas com o shape especificado. Pode ser mais rápido que zeros/ones para arrays grandes onde os valores não importam imediatamente.
*   [`np.eye(N)`](https://numpy.org/doc/stable/reference/generated/numpy.eye.html): Cria uma matriz identidade de dimensão `N x N` (diagonal principal com uns, o restante zeros).
*   [`np.diag(v)`](https://numpy.org/doc/stable/reference/generated/numpy.diag.html): Cria uma matriz diagonal com os elementos de `v` na diagonal principal, ou extrai a diagonal de uma matriz.
*   [`np.linspace(start, stop, num)`](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html): Cria um array com `num` elementos igualmente espaçados entre `start` e `stop` (inclusive por padrão).
*   [`np.random.rand(d0, d1, ..., dn)`](https://numpy.org/doc/stable/reference/random/generated/numpy.random.rand.html): Cria um array com valores aleatórios uniformemente distribuídos entre 0 e 1.
*   [`np.random.randint(low, high, size)`](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html): Cria um array com inteiros aleatórios no intervalo `[low, high)`.
*   [`np.fromfunction(function, shape)`](https://numpy.org/doc/stable/reference/generated/numpy.fromfunction.html): Aplica uma função a cada coordenada para gerar valores.


````python
import numpy as np

# Exemplo de Criação de Arrays Específicos

# Array de zeros (2x3)
zeros = np.zeros((2, 3))
print(f"Array de zeros:\n{zeros}")

# Array de uns (3x2)
ones = np.ones((3, 2))
print(f"\nArray de uns:\n{ones}")

# Array preenchido com um valor específico (2x2 com 7)
full = np.full((2, 2), 7)
print(f"\nArray preenchido com 7:\n{full}")

# Matriz identidade (3x3)
identidade = np.eye(3)
print(f"\nMatriz Identidade:\n{identidade}")

# Matriz diagonal a partir de um vetor
diagonal_vec = np.diag([1, 2, 3])
print(f"\nMatriz Diagonal (a partir de vetor):\n{diagonal_vec}")

# Valores igualmente espaçados (0 a 10, 5 elementos)
linspace = np.linspace(0, 10, 5)
print(f"\nLinspace (0 a 10, 5 elementos): {linspace}")

# Números aleatórios (uniforme 0-1)
random_uniform = np.random.rand(2, 2)
print(f"\nNúmeros aleatórios (uniforme 0-1):\n{random_uniform}")

# Números inteiros aleatórios (1 a 10, 3 elementos)
random_int = np.random.randint(1, 10, size=3)
print(f"Números inteiros aleatórios (1 a 9): {random_int}")

# Criação a partir de função
def func(i, j):
    return i + j
arr_func = np.fromfunction(func, (3, 3), dtype=int)
print(f"\nArray criado a partir de função (i+j):\n{arr_func}")
````
