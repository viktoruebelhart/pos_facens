# Parte 1: Mergulho no Matplotlib

O Matplotlib é a fundação de quase toda a visualização em Python. Ele foi projetado para se assemelhar aos comandos de plotagem do MATLAB.

- **Arquitetura de Camadas**: O Matplotlib possui três camadas principais: a `Backend` (que renderiza o gráfico), a `Artist` (onde você define como o gráfico deve parecer) e a `Scripting` (`pyplot`).
- **Interface Orientada a Objetos**: Diferente da interface simples, a OO permite que você manipule objetos `Figure` (o container total) e `Axes` (a área do gráfico) de forma independente. Isso é crucial para criar layouts complexos ou modificar detalhes específicos após a criação do gráfico.
- **Customização Granular**: Você tem controle total. Se precisar mudar a cor de apenas um traço específico no eixo X ou criar uma anotação com seta em uma coordenada exata, o Matplotlib é a ferramenta.
- **Quando usar**: Para gráficos altamente customizados, publicações científicas que exigem padrões rigorosos e quando o Seaborn não oferece o layout exato que você deseja.

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Preparando dados
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 2. Criando a Figura e os Eixos (Interface Orientada a Objetos)
fig, ax = plt.subplots(figsize=(10, 6))

# 3. Plotagem principal
ax.plot(x, y, label='Senoide', color='blue', linestyle='--', linewidth=2)

# 4. Customização de Eixos e Títulos
ax.set_title('Exemplo de Funções Matplotlib', fontsize=15)
ax.set_xlabel('Eixo X (Tempo)')
ax.set_ylabel('Eixo Y (Amplitude)')

# 5. Adicionando Grade e Legenda
ax.grid(True, linestyle=':', alpha=0.6)
ax.legend(loc='upper right')

# 6. Anotação
ax.annotate('Pico Máximo', xy=(1.57, 1), xytext=(3, 1.2),
             arrowprops=dict(facecolor='black', shrink=0.05))

# 7. Exibição
plt.show()
```

### Variações de Gráficos com Matplotlib
Vamos explorar como criar gráficos de barras, pizza e dispersão usando a mesma interface.

```python
import matplotlib.pyplot as plt

# --- 1. Gráfico de Barras ---
plt.figure(figsize=(8, 5))
categorias = ['Produto A', 'Produto B', 'Produto C']
valores = [25, 40, 15]

plt.bar(categorias, valores, color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title('Comparativo de Vendas por Produto')
plt.ylabel('Quantidade Vendida')
plt.show()
```

```python
import matplotlib.pyplot as plt

# --- 2. Gráfico de Pizza ---
plt.figure(figsize=(6, 6))
labels = ['Android', 'iOS', 'Windows', 'Outros']
tamanhos = [45, 30, 15, 10]
explode = (0.1, 0, 0, 0)  # 'Explodir' a primeira fatia

plt.pie(tamanhos, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=140, colors=plt.cm.Pastel1.colors)
plt.title('Distribuição de Sistemas Operacionais')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- 3. Gráfico de Dispersão (Scatter Plot) ---
x = np.random.rand(50)
y = np.random.rand(50)
cores = np.random.rand(50)
tamanhos = 1000 * np.random.rand(50)

plt.figure(figsize=(8, 5))
plt.scatter(x, y, c=cores, s=tamanhos, alpha=0.5, cmap='viridis')
plt.title('Relação entre Variáveis Aleatórias')
plt.colorbar(label='Intensidade')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Subplots e Anotações ---
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Histograma no primeiro subplot
dados = np.random.randn(1000)
ax1.hist(dados, bins=30, color='skyblue', edgecolor='black')
ax1.set_title('Histograma')

# Dispersão com Anotação no segundo subplot
x_val, y_val = [1, 2, 3], [10, 20, 15]
ax2.scatter(x_val, y_val, color='red')
ax2.annotate('Ponto Importante', xy=(2, 20), xytext=(2.5, 22),
             arrowprops=dict(facecolor='black', shrink=0.05))
ax2.set_title('Dispersão com Anotação')

plt.tight_layout()
plt.show()
```

```python
# --- 2. Eixo Duplo (Twinx) ---
fig, ax1 = plt.subplots(figsize=(8, 5))

x = np.linspace(0, 10, 100)
ax1.plot(x, np.exp(x/3), 'g-', label='Crescimento Exp')
ax1.set_ylabel('Escala Exponencial', color='g')

# Criando o segundo eixo Y
ax2 = ax1.twinx()
ax2.plot(x, np.sin(x), 'b--', label='Oscilação Seno')
ax2.set_ylabel('Escala Senoidal', color='b')

plt.title('Gráfico de Eixo Duplo')
plt.show()
```

# Parte 2: A Magia do Seaborn

O Seaborn é uma biblioteca de alto nível construída sobre o Matplotlib. Sua principal filosofia é tornar a visualização de dados uma parte central da exploração e compreensão de dados estatísticos.

- **Mapeamento Semântico**: Permite associar colunas dos dados a propriedades visuais (cor, tamanho, estilo) de forma declarativa.
- **Agregação e Estimativa**: Gráficos como `sns.lineplot` calculam automaticamente médias e intervalos de confiança.
- **Paletas de Cores**: Oferece sistemas de cores baseados na percepção humana (como `magma` ou `viridis`).
- **Quando usar**: Para análise exploratória rápida, mapas de calor (`heatmap`) e distribuições complexas.

### Exemplos Práticos com Seaborn

Vamos ver como o Seaborn facilita a criação de gráficos complexos com poucas linhas de código utilizando os datasets que carregamos.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configuração global de estilo
sns.set_theme(style="whitegrid")

# 1. Distribuição: Histograma com KDE (Kernel Density Estimate)
plt.figure(figsize=(10, 5))
sns.histplot(data=titanic, x='Age', hue='Survived', kde=True, palette='viridis')
plt.title('Distribuição de Idade por Sobrevivência')
plt.show()
```

```python
# 2. Relacional: Scatter Plot com Regressão
plt.figure(figsize=(10, 6))
sns.scatterplot(data=titanic, x='Age', y='Fare', hue='Pclass', size='Fare', sizes=(20, 200))
plt.title('Relação Idade vs Tarifa por Classe')
plt.show()
```

```python
# 3. Categórico: Boxplot e Swarmplot combinados
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic, x='Embarked', y='Age', palette='Set2')
sns.swarmplot(data=titanic, x='Embarked', y='Age', color=".25", size=3)
plt.title('Distribuição de Idade por Porto de Embarque')
plt.show()
```

```python
# 4. Matriz: Heatmap de Correlação
plt.figure(figsize=(8, 6))
# Selecionando apenas colunas numéricas para correlação
corr = titanic.select_dtypes(include=['number']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de Calor de Correlação (Titanic)')
plt.show()
```

```python
# --- 1. Regressão e Densidade ---
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Densidade (KDE)
sns.kdeplot(data=titanic, x='Age', hue='Sex', fill=True, ax=ax[0])
ax[0].set_title('Densidade de Idade por Sexo')

# Regressão Linear (lmplot usa sua própria figura, aqui usamos regplot para o subplot)
sns.regplot(data=titanic, x='Age', y='Fare', ax=ax[1], scatter_kws={'alpha':0.3})
ax[1].set_title('Regressão: Idade vs Tarifa')

plt.show()
```

```python
# --- 2. Contagens e Grades (FacetGrid / PairPlot) ---
# Countplot Horizontal
plt.figure(figsize=(8, 4))
sns.countplot(data=netflix.head(100), y='type', palette='pastel')
plt.title('Contagem de Títulos por Tipo (Amostra)')
plt.show()

# Exemplo de PairPlot (usando amostra do Titanic)
sns.pairplot(titanic[['Survived', 'Pclass', 'Age', 'Fare']].dropna(), hue='Survived', palette='bright')
plt.show()
```
