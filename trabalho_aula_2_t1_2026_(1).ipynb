## Exercício 1

Crie uma lista de números inteiros. Em seguida, utilize a função `map` (sem loops explícitos) para criar uma nova lista onde cada número da lista original é multiplicado por 2. Por fim, imprima a nova lista.

```python
# Digite seu código aqui
lista = [1, 2, 3, 4, 5]
nova_lista = list(map(lambda x: x * 2, lista))
print(nova_lista)
```

## Exercício 2

Dada uma lista de palavras, utilize a função `filter` (sem loops explícitos) para criar uma nova lista contendo apenas as palavras que têm mais de 5 caracteres. Imprima a lista filtrada.

```python
# Digite seu código aqui
palavras = ["oi", "gato", "elefante", "passarinho", "tartaruga"]
palavras_5 = list(filter(lambda x: len(x) > 5, palavras))

print("lista de palavras:", palavras)
print("lista de palavras com mais de 5 caracteres:", palavras_5)
```

## Exercício 3

Usando `functools.reduce`, calcule o produto de todos os números em uma lista de inteiros. Você precisará importar `reduce` de `functools`. Imprima o resultado.

```python
# Digite seu código aqui
from functools import reduce
lista = [1, 2, 3, 4, 5, 10, 2]
produto = reduce(lambda x, y: x * y, lista)
print(produto)
```

## Exercício 4

Dada uma lista de números, use uma combinação de `filter` e `map` (e funções `lambda` anônimas) para primeiro filtrar apenas os números pares e, em seguida, elevar cada um desses números pares ao quadrado. Imprima a lista resultante.

```python
# Digite seu código aqui
lista = [1, 2, 3, 4, 5, 10, 2]
pares = list(filter(lambda x: x % 2 == 0, lista))
print("Lista de Pares: ", pares)
ao_quadrado = list(map(lambda x: x**2, pares))
print("Lista de Pares²: ", ao_quadrado)
```
